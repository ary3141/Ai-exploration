# Import The Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Import the Data

```python
databayam = pd.read_csv('/Users/mdwivaaryaerlangga/Downloads/Position_Salaries.csv')
X = databayam.iloc[:,1:-1].values
y = databayam.iloc[:, -1].values
```

# Training The Decisiton tree Regression on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state= 0)
regressor.fit(X,y)
```

# Predicting a new result

```python
regressor.predict([[6.5]])
```

# Visualizing the Decision Tree regression (for higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.001)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Decision Tree Regression (Truth Or Bluff)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

# Visualizing the Polynomial Linear regression model

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Decision Tree (Truth Or Bluff)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```
