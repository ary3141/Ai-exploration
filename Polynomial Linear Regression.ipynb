# Import The Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Import the Data

```python
databayam = pd.read_csv('/Users/mdwivaaryaerlangga/Downloads/Position_Salaries.csv')
X = databayam.iloc[:,1:-1].values
y = databayam.iloc[:, -1].values
```

# Training The Linear Regression on the whole dataset

```python
from sklearn.linear_model import LinearRegression
regressor_linear = LinearRegression()
regressor_linear.fit(X,y)
```

# Training The Polynomial Linear Regression on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
regressor_polynomial = PolynomialFeatures(degree = 4)
X_poly = regressor_polynomial.fit_transform(X)
regressor_linear_2 = LinearRegression()
regressor_linear_2.fit(X_poly, y)
```

# Visualizing the linear regression model

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, regressor_linear.predict(X), color = 'blue')
plt.title('Linear Regression (Truth Or Bluff)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

# Visualizing the Polynomial Linear regression model

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, regressor_linear_2.predict(X_poly), color = 'blue')
plt.title('Polynomial Regression (Truth Or Bluff)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

# Visualizing the polynomial Linear regression (for higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y, color = 'red')
plt.plot(X_grid, regressor_linear_2.predict(regressor_polynomial.fit_transform(X_grid)), color = 'blue')
plt.title('Polynomial Regression (Truth Or Bluff)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

# Predicting a new result with linear regression

```python
regressor_linear.predict([[6.5]])
```

# Predicting a new result with polynomial linear regression

```python
regressor_linear_2.predict(regressor_polynomial.fit_transform([[6.5]]))
```
